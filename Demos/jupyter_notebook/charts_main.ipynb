```python
def data_gen():
    pipeline = [
#            {'$match':{table2:"湖南省"}},
#            {'$match':{table3:"已获投资"}},
           {'$group':{'_id':tables,'counts':{'$sum':1}}},                                     #group命令是用来分组
           {'$sort' :{'counts':-1}},
           {'$limit': 25} 
    ]
    for i in mongo_table.aggregate(pipeline):
        yield [i['_id'],i['counts']]

count1=[]
count2=[]
for i,b in enumerate(data_gen()):
    count1.append(b[0])
    count2.append(b[1])
    
    print(i+1,b[0])
```

```python
def data_gen(types):
    length = 0
    if length <= len(count1):
        for name,data in zip(count1,count2):
            datas = {
            'name':name,
            'data':[data],
            'type':types
            }
            yield datas
            length += 1

            
for i in data_gen('column'):
    print(i)
```

```python
titlet = "2018年全国 “互联网+”大学生创新创业大赛 高校创赛项目（前25） 统计"
series = [datas for datas in data_gen('column')]
charts.plot(series,options=dict(title=dict(text=titlet)),show='inline')
```
