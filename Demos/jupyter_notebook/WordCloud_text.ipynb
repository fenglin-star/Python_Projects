```python
import pymongo
import matplotlib.pyplot as plt
from string import punctuation 
from wordcloud import WordCloud
import jieba
```

```python
# 写入参数
#通用参数
titlet = '智联招聘职位数据_学前教育'                #'输入你图表的标题'                                  
mongodb_db = 'Copy'                                #'输入你数据库名'
mongodb_table = '学前教育'                         #'输入你数据的表格名'

#柱形图参数
table = 'brief'                                   #'输入你查询数据的字典名'

#饼图参数
text = '招聘薪水最多的前12位'                     #'输入你图表的摘要'
tables = '$' + table                              #'输入你查询数据的字典名'
```

```python
client = pymongo.MongoClient('localhost',27017)
mongodb_db = client[mongodb_db]
mongo_table  = mongodb_db[mongodb_table]
```

```python
MAX=[]
for i in mongo_table.find()[:20]:
    MAX.append(i[table])
#     print(i)
max=''.join(MAX)
print(max)
```

```python
# -*-coding:utf-8-*-

###生成txt文件的词云

# 添加的自定义中文语句的代码在这里
jieba.add_word('工作经验')
jieba.add_word('学前教育')
jieba.add_word('团队合作')
jieba.add_word('专业优先')
jieba.add_word('热爱')
# jieba.add_word('团队合作')

#停用词
STOPWORDS=""

text = max
# 结巴分词
wordlist = jieba.cut(text, cut_all=False)
wl = " ".join(wordlist)
# print(wl)#输出分词之后的txt

# # 把分词后的txt写入文本文件
# fenciTxt  = open("fenciHou.txt","w+")
# fenciTxt.writelines(wl)
# fenciTxt.close()

# 使用 jieba 清理停用词
def jiebaclearText(text):
    mywordlist = []
    seg_list = jieba.cut(text, cut_all=False)
    liststr = "/ ".join(seg_list)
    STOPWORD = jieba.cut(STOPWORDS, cut_all=False)
    w2 = " ".join(STOPWORD)
    f_stop_seg_list = w2
    print(f_stop_seg_list)
    for myword in liststr.split('/'):
        if not (myword.strip() in f_stop_seg_list) and len(myword.strip()) > 1:
            mywordlist.append(myword)
    return ''.join(mywordlist)

findlly_text = jiebaclearText(text)
print(findlly_text)#输出分词之后的txt/


# 设置词云
wc = WordCloud(background_color="black",  # 设置背景颜色
               # mask = "图片",  #设置背景图片
               max_words=200,  # 设置最大显示的字数
#                stopwords = STOPWORDS, #设置停用词
               font_path="C:\Windows\Fonts\AdobeFangsongStd-Regular.otf",
               # 设置中文字体，使得词云可以显示（词云默认字体是“DroidSansMono.ttf字体库”，不支持中文）
               max_font_size=50,  # 设置字体最大值
               random_state=50,  # 设置有多少种随机生成状态，即有多少种配色方案
#                width=2000, height=1860, margin=2,  # 设置图片默认的大小,但是如果使用背景图片的话,那么保存的图片大小将会
               )

myword = wc.generate(findlly_text)  # 生成词云

# 展示词云图
plt.imshow(myword)
plt.axis("off")
plt.show()
```
