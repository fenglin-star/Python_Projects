```python
#coding:utf-8
import pymongo
from string import punctuation 
import charts
```

```python
titlet = '教育学历分布'   #'输入你图表的标题'
mongodb_db = '湖南第一师范学院_招聘情况统计'      #'输入你数据库名'
mongodb_table = '湖南第一师范学院_校园招聘'          #'输入你数据的表格名'
table = 'city_name '                    #'输入你查询数据的字典名'
client = pymongo.MongoClient('localhost',27017)
mongodb_db = client[mongodb_db]
mongo_table  = mongodb_db[mongodb_table]
```

```python
for i in mongo_table.find()[:20]:
    print(i[table])
```

```python
for i in mongo_table.find():
    if i['city_name ']:
        area = [i for i in i['city_name '] if i not in punctuation]
    else:
        area = ['不明']
    area = ''.join(area)
    mongo_table.update({'_id': i['_id']},{'$set':{table: area}}, True)
```

```python
    area_list = []
    for i in mongo_table.find():
        area_list.append(i[table])
    area_index = list(set(area_list))
    print(area_index)
```

```python
post_times = []
for index in area_index:
    post_times.append(area_list.count(index))
print(post_times)
```

```python
def data_gen(types):
    length = 0
    if length <= len(area_index):
        for name,data in zip(area_index,post_times):
            datas = {
            'name':name,
            'data':[data],
            'type':types
            }
            yield datas
            length += 1
```

```python
for i in data_gen('column'):
    print(i)
```

```python
series = [datas for datas in data_gen('column')]
charts.plot(series,options=dict(title=dict(text=titlet)),show = 'inline')
```
